```python
file = r"C:\Users\mwver\OneDrive\Documents\GitHub\Machine-Learning-and-Neural-Networks-Practice\iris_dataset\iris.csv"
df = pd.read_csv(file)
df.insert()
df.describe()
```

```python
color_classes = {
    "Setosa":"red",
    "Versicolor":"blue",
    "Virginica":"green"
}

def plotScatter(x, y):

    for class_name, color in color_classes.items():
        class_df = df[df['class'] == class_name]
        plt.scatter(class_df[x], class_df[y], color=color, label=class_name)

    plt.xlabel('X-axis Label')
    plt.ylabel('Y-axis Label')
    plt.title('Scatter Plot with Different Colors for Classes')
    plt.legend()
    plt.show()

```

```python
plotScatter("sepal length", "sepal width")
plotScatter("petal length", "petal width")
plotScatter("sepal length", "petal width")
plotScatter("petal length", "sepal width")
```
